```python

### A. Discounts and Pricing

# 1. Check Discount Eligibility
purchase_amount = float(input("Enter total purchase amount: $"))
if purchase_amount > 100:
    final_price = purchase_amount * 0.9
else:
    final_price = purchase_amount
print("Final price after discount (if applicable): $", final_price)

# 2. Calculate Bulk Discount
items = int(input("Enter the number of items purchased: "))
total_price = float(input("Enter total price: $"))
if items > 5:
    total_price *= 0.85
print("Total price after bulk discount (if applicable): $", total_price)

# 3. Membership Discount
is_member = input("Are you a member (yes/no)? ").lower() == 'yes'
price = float(input("Enter total price: $"))
if is_member:
    price *= 0.8
else:
    price *= 0.95
print("Price after membership discount: $", price)

# 4. Seasonal Sale
is_holiday = input("Is today a holiday (yes/no)? ").lower() == 'yes'
price = float(input("Enter price: $"))
if is_holiday:
    price *= 0.75
else:
    price *= 0.9
print("Price after seasonal discount: $", price)

# 5. Buy-One-Get-One-Free
items = int(input("Enter the number of items purchased: "))
pay_items = items // 2 + items % 2
print("Number of items to pay for:", pay_items)
```

```python
### B. Tax Calculations

# 6. Sales Tax
price = float(input("Enter the price of the item: $"))
if price > 500:
    price *= 1.15
else:
    price *= 1.08
print("Price after tax: $", price)

# 7. Income Tax
income = float(input("Enter annual income: $"))
if income > 50000:
    tax = income * 0.2
else:
    tax = income * 0.1
print("Tax amount: $", tax)

# 8. Tax Bracket
income = float(input("Enter annual income: $"))
if income < 30000:
    bracket = "Low Tax"
elif income < 100000:
    bracket = "Medium Tax"
else:
    bracket = "High Tax"
print("Tax Bracket:", bracket)

# 9. VAT Calculation
is_essential = input("Is the item essential (yes/no)? ").lower() == 'yes'
price = float(input("Enter price of the item: $"))
if is_essential:
    price *= 1.05
else:
    price *= 1.12
print("Price after VAT: $", price)

# 10. Tax-Free Day
tax_free = input("Is today a tax-free day (yes/no)? ").lower() == 'yes'
price = float(input("Enter the original price: $"))
if not tax_free:
    price *= 1.07
print("Final price: $", price)
```

```python
### C. Shopping and Billing

# 11. Free Shipping
purchase_amount = float(input("Enter total purchase amount: $"))
if purchase_amount > 50:
    shipping = 0
else:
    shipping = 5
total = purchase_amount + shipping
print("Total amount including shipping: $", total)

# 12. Discount Code
price = float(input("Enter total price: $"))
discount_code = input("Enter discount code: ")
if discount_code == "DISCOUNT10":
    price *= 0.9
print("Final price after discount (if applicable): $", price)

# 13. Tiered Discounts
price = float(input("Enter total price: $"))
if price <= 50:
    discount = 0
elif price <= 100:
    discount = 0.1
else:
    discount = 0.2
final_price = price * (1 - discount)
print("Final price after tiered discount: $", final_price)

# 14. Minimum Purchase Requirement
amount = float(input("Enter total amount: $"))
if amount < 20:
    print("Minimum purchase of $20 is required.")
else:
    print("Total amount: $", amount)

# 15. Loyalty Points
is_loyal = input("Is the customer a loyal member (yes/no)? ").lower() == 'yes'
purchase_amount = float(input("Enter purchase amount: $"))
points = purchase_amount * 2 if is_loyal else purchase_amount
print("Loyalty points earned:", points)
```

```python
### D. Travel and Tickets

# 16. Travel Discount
distance = float(input("Enter travel distance in miles: "))
ticket_price = float(input("Enter ticket price: $"))
if distance > 500:
    ticket_price *= 0.8
print("Final ticket price: $", ticket_price)

# 17. Child or Senior Discount
age = int(input("Enter passenger's age: "))
ticket_price = float(input("Enter ticket price: $"))
if age < 12 or age > 60:
    ticket_price *= 0.85
print("Final ticket price: $", ticket_price)

# 18. Ticket Type Pricing
is_weekend = input("Is the ticket for a weekend (yes/no)? ").lower() == 'yes'
ticket_price = float(input("Enter ticket price: $"))
if is_weekend:
    ticket_price *= 1.1
print("Final ticket price: $", ticket_price)

# 19. Baggage Fee
baggage_weight = float(input("Enter total baggage weight (kg): "))
if baggage_weight > 20:
    extra_fee = (baggage_weight - 20) * 10
else:
    extra_fee = 0
print("Extra baggage fee: $", extra_fee)

# 20. Early Bird Discount
booking_days = int(input("Enter number of days before travel: "))
ticket_price = float(input("Enter ticket price: $"))
if booking_days > 30:
    ticket_price *= 0.9
print("Final ticket price: $", ticket_price)
```

```python
# E. Grades and Performance

# 21. Pass or Fail
score = int(input("Enter student's score: "))
if score >= 40:
    print("Pass")
else:
    print("Fail")

# 22. Grade Assignment
score = int(input("Enter student's score: "))
if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 50:
    grade = "C"
else:
    grade = "F"
print("Grade:", grade)

# 23. Bonus Marks
score = int(input("Enter student's score: "))
completed_assignments = input("Did the student complete all assignments (yes/no)? ").lower() == 'yes'
if completed_assignments:
    score += 5
print("Final score with bonus (if applicable):", score)

# 24. Attendance Eligibility
attendance = float(input("Enter student's attendance percentage: "))
if attendance >= 75:
    print("Eligible to take the exam.")
else:
    print("Not eligible to take the exam.")

# 25. Scholarship Eligibility
grade = input("Enter student's grade: ").upper()
family_income = float(input("Enter annual family income: $"))
if grade == "A" and family_income < 30000:
    print("Eligible for scholarship.")
else:
    print("Not eligible for scholarship.")
```
